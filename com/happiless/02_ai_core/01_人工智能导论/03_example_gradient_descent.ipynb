```python
import numpy as np
import matplotlib.pyplot as plt
import random

def func(x):
    return 10 * x ** 2 + 32 * x + 9

def gradient(x):
    return 20 * x + 32
```

```python
x = np.linspace(-10, 10)
steps = []
x_star = random.choice(x)
alpha = 1e-3
```

```python
x_star
```

```python
x
```

```python

for i in range(100):
    x_star = x_star - gradient(x_star) * alpha
    steps.append(x_star)
    plt.scatter(x, func(x))
    plt.scatter(x_star, func(x_star))
    print(x_star, func(x_star))
    plt.show()
```

```python

fig, ax = plt.subplots()
ax.plot(x, func(x))
for i, s in enumerate(steps):
    ax.annotate(str(i+1), (s, func(s)))
plt.show()
```
