### 1. 统计学习
-   1. 统计学习的特点: 关于计算机基于数据构建概率统计模型并运用模型对数据进行预测与分析的一门学科
-   2. 统计学习对象: 数据(data), 从数据出发，提取数据的特征， 抽象出数据的模型，又回到对数据的分析和预测中去
-   3. 统计学习目的: 对数据进行预测和分析
-   4. 统计学习方法:
        1. 监督学习，非监督学习，半监督学习，强化学习
        2. 三要素: 模型，策略，算法
-   5. 步骤:<br>
        1). 得到一个有限的训练数据集合；<br>
        2). 确定包含所有可能的模型的假设空间，即学习模型的集合；<br>
        3). 确定模型选择的准则，即学习的策略；<br>
        4). 实现求解最优模型的算法，即学习的算法；<br>
        5). 通过学习方法选择最优模型；<br>
        6). 利用学习的最优模型对新数据进行预测或分析<br>
-   6. 统计学习的研究: 统计学习方法， 统计学习理论， 统计学习应用
-   7. 统计学习的重要性<br>
        1). 处理海量数据的有效手段<br>
        2). 实现计算机智能化的有效手段<br>
        3). 计算机科学的重要组成部分<br>

### 1.2 监督学习
-   1. 输入空间、特征空间和输出空间<br>
        输入空间，输出空间: 输入和输出所有可能的取值的集合<br>
        特征空间: 每个输入是一个实例，通常由特征向量组成，特征向量存在的空间<br>
-   2. 联合概率分布: 监督学习假设X和Y遵循联合概率分布<br>
-   3. 假设空间: 模型属于由输入空间到输出空间的映射的集合<br>

        模型f(x)关于训练集的平均损失称为经验风险或经验损失:
    $ R_{emp}(f) = \frac{1}{N}\sum\limits_{i=1}^{N}L(y_i, f(x_i))$
-   2. 经验风险最小化与结构风险最小化
        1). 在假设空间，损失函数以及训练数据确定的情况下，经验风险可以确定，经验风险最小的模型就是最优的模型<br>
        2). 正则项: 当样本比较少时，容易出现过拟合现象,所以给以上的经验风险函数加上惩罚项<br>
    $ R_{emp}(f) = \frac{1}{N}\sum\limits_{i=1}^{N}L(y_i, f(x_i)) + \lambda J(f)$
    其中$J(f)$为模型的复杂度, $\lambda$ 为常数
